```python
import pandas as pd
from pathlib import Path
```

```python
raw_path = Path("../data/raw")

csv_files = sorted(raw_path.glob("*.csv"))

print(f"Found {len(csv_files)} files")
```

```python
for file in csv_files:

    df = pd.read_csv(file)

    print("\n" + "="*50)
    print(file.name)

    print("\nShape:")
    print(df.shape)

    print("\nColumns:")
    print(df.columns.tolist())
```

```python
for file in csv_files:

    df = pd.read_csv(file)

    print("\n" + "="*50)
    print(file.name)

    print("\nData Types:")
    print(df.dtypes)
```

```python
for file in csv_files:

    df = pd.read_csv(file)

    print("\n" + "="*50)
    print(file.name)

    print("\nMissing Values:")
    print(df.isnull().sum())
```

```python
fund_master = pd.read_csv(
    "../data/raw/01_fund_master.csv"
)

nav_history = pd.read_csv(
    "../data/raw/02_nav_history.csv"
)
```

```python
fund_master_codes = set(fund_master["amfi_code"])

nav_history_codes = set(nav_history["amfi_code"])

missing_codes = fund_master_codes - nav_history_codes

print("Missing Codes:")
print(missing_codes)
```

```python
print("Fund Master Codes:", len(fund_master_codes))
print("NAV History Codes:", len(nav_history_codes))
print("Missing Codes:", len(missing_codes))
```
